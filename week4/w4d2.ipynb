# Cleaning data

```python
!uchardet baby_names.csv
```

```python
genders = ('female', 'male')
ranks = ('1', '2', '3', '4', '5')
columns = [(gender, rank) for gender in genders for rank in ranks]
print(columns)
index = pd.MultiIndex.from_tuples(columns)
index
```

```python
baby_names = pd.read_csv('baby_names.csv', encoding='windows-1252',
                         header=9, index_col=0, names=index)
baby_names.head(10)
```

```python
baby_names = baby_names[:-3]
```

```python
baby_names['female'].head()
```

```python
help(baby_names.stack)
```

```python
baby_names.stack().head(20)
```

```python
baby_names.stack(level=0).head(20)
```

```python
baby_names.stack(level=0).stack(level=0)
```

`stack()` takes a header row (column labels) and moves them to the index. If this is the last header row, then the dataframe becomes a series.

**You can't `stack()` a series!**

```python
baby_names[:5]
```

```python
baby_names = baby_names.stack(level=0)
```

```python
baby_names
```

```python
# it is possible to undo a stack

baby_names.unstack().stack(level=0).unstack()
```

```python
baby_names.head()
```

```python
baby_names.xs('2013')
```

```python
help(baby_names.xs)
```

We can combine cross section levels together

```python
baby_names.xs(['2013', 'female'])['3']
```
