```python
# convert the numeric columns to floats from strings

numeric_columns = ['1990 [YR1990]', '1995 [YR1995]',
                   '2000 [YR2000]', '2005 [YR2005]',
                   '2010 [YR2010]', '2013 [YR2013]',
                   '2014 [YR2014]', '2015 [YR2015]']
for col in numeric_columns:
    wdi_data[col] = wdi_data[col].map(float)
```
