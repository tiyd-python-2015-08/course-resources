# Indexing by date

A lot of data has some sort of timestamp associated with it

One common operation is **resampling** by timestamp - that is, taking a large set of data and either shrinking it down and summarizing it by date, or even expanding the dates and filling in blanks to get more spread out data

All of this requires us to be aware of the dates as actual timestamps, though

Because timestamps can take a lot of different forms, pandas will not automatically identify/convert them. We must manually tell it how to interpret dates in our data

```python
rdu_weather['DATE'].head()
```

We can pull out the `DATE` column, but we still need to convert to a timestamp. Also, we can convert to an **index**

We **index** our data to pull unique items out quickly.

Indexing also lets us use pandas features like _resampling_.

So what we do is **remove** the DATE column from our dataframe, and **reapply** as an index. If we don't remove the DATE column, it will be in our data twice, which is redundant and also gives opportunities for us to get them out of sync

```python
date_series = rdu_weather.pop('DATE')
date_series.head()
```

```python
rdu_weather.head()
```

```python
# putting the column back in as an INDEX

rdu_weather.index = pd.to_datetime(date_series, format='%Y%m%d')
rdu_weather.head()
```

# Cleaning up Data

The reality of life is that most data is dirty

Data scientists spend 90% of their time cleaning up data

For this weather data, the creator of data has decided that -9999 represents "no data available"

In the case of this data, it makes sense - it's something out of our range of expected values. So we have to deal with their hacks

One way is using `.replace()`

### replace()

NaN (**not a number**) is commonly used to represent something that's not a number inside a field that is supposed to be a float

_(sidenote: Python 3.5 introduces math.nan)_

For older versions of Python, use `np.nan`

We can use the `.replace()` method combined with `np.nan` to find/replace all -9999 values with nan

Let's try it on a small set of data first to make sure it works

```python
rdu_weather.replace(-9999, np.nan)[0:4].describe()
```

```python
rdu_weather.head()
```

The `.replace()` method of DataFrame objects works just like the `.replace()` of strings - it doesn't change the DataFrame object, it returns a copy of it with the changes made. That's be design.

One way of dealing with that is to replace the object itself with the copy returned by `.replace()`

```python
rdu_weather = rdu_weather.replace(-9999, np.nan)
rdu_weather.head()
```

```python
rdu_weather.describe()
```

So using replace() and = actually replaces our DataFrame object.

## Unit conversion

Our temperatures are stored as integers representing tenths of degrees C

Let's write a function that will convert our temperature data into a float that's readable by us American types



```python
def c10_to_f(temp):
    return ((temp / 10) * (9 / 5) + 32).round(1)

rdu_temps['temp_max'] = c10_to_f(rdu_temps.pop('TMAX'))
```

```python
rdu_temps['temp_min'] = c10_to_f(rdu_temps.pop('TMIN'))
```

```python
rdu_temps.head()
```

```python
(rdu_weather['PRCP'] - rdu_weather['SNOW'])
```

### In-place conversion

You don't need to write a function to do conversions/manipulations - you can just write a simple expression

* wind speed is in 10ths of meters per second - convert to m/s
* precipitation is in 10ths of mm - convert to mm
* snowfall is in millimeters - that's fine, we should rename it to be more readable

```python
rdu_temps['wind_speed'] = rdu_temps.pop('AWND') / 10
rdu_temps['precipitation'] = rdu_temps.pop('PRCP') / 10
```

```python
rdu_temps.head()
```

```python
rdu_temps.rename(columns={'SNOW': 'snowfall'}).head()
```

`.rename()` works like `.replace()` - it just returns a copy, it doesn't change the original data frame

If we want to change the DataFrame, we need to reasssign it, just like with strings

```python
rdu_temps = rdu_temps.rename(columns={'SNOW': 'snowfall'})
rdu_temps.head()
```

## Querying Data

There are a few different ways to run queries using pandas. One really cool way is using what's known as **Boolean Indexing**. Basically, we pass an expression that will evaluate to True or False inside the square brackets

For instance, let's find days taht had snowfall.

```python
rdu_temps[rdu_temps.snowfall > 0]
```

```python
rdu_temps.snowfall > 0
```

**Boolean Indexing** works by creating a **series** of Boolean values assocaited with an index and passing those into the square brackets of the DataFrame

# Time series manipulation

**.resample()**

When we want to manipulate data to focus on a different _sampling_ of the indexes, we can use `resample` to do that.

`resample` takes two main parameters:

* Rule ('Time aggregate')
* How

The **Rule** lets us define how detailed we want the output (monthly, or weekly, or hourly). If we are asking for a higher resolution of resampling than was originally in the data, it has to make up data. The default made up data is NaN.

The **How** lets us define what data will be stuck in each row. That can just be the last data we saw, or we can do aggregate functions, like `sum` or `mean` or `stdev`

```python
rdu_temps.precipitation.resample('M', how=['sum', 'mean'])
```

```python
rdu_temps.temp_max.resample('w', how=['mean', 'median', 'std', 'count'])
```

If you only provide a **rule** to the `resample` method, then it will only show the row representing the closest match to how you've defined the rule. "M" is the end of the month

```python
rdu_temps.resample('M')
```

```python
rdu_temps.temp_max.resample('W', how=['mean', 'median', 'std'])
```

```python
rdu_temps[['temp_max', 'temp_min']].resample('M', ['mean', 'median', 'std'])
```

```python
rdu_temps[['temp_max', 'temp_min']].resample('M', ['mean', 'median', 'std']).stack()
```

# Plotting

Resampling can especially be useful for plotting, rather than plotting every single point

```python
rdu_temps.temp_max.resample('MS', how=['mean', 'std']).plot(title='Monthly average max temp')
rdu_temps.temp_max.resample('W', how=['mean', 'std']).plot(title='Weekly average max temp')
```

```python
rdu_temps.temp_max.plot(title='Daily max temp')
```

```python
rdu_temps[['temp_max', 'temp_min']].resample('W', ['mean']).plot(title='Weekly highs/lows')
```
